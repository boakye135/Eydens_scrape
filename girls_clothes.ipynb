```python
""" I wrote this code to scrape my friends business page to collect information on the name of products
and their respective prices. The website had a total of 22 pages."""
import requests,bs4
import pandas as pd
```

```python
rows = []
columns = ['product name','product cost']
for x in range(1,23):#total of 22 pages
    r = requests.get(f'https://www.eydens.com/shop/page/{x}/')
    soup = bs4.BeautifulSoup(r.content,'html.parser')
    products = soup.findAll('ul',{'class':'woo-entry-inner clr'})
    for product in products:
        product_name = product.find('li',{'class':'title'}).text.strip()
        product_cost = product.find('span',{'class':'woocs_price_code'}).text.strip()
        row = [product_name,product_cost]
        rows.append(row)
        print(product_name + ' costs ' + str(product_cost) + '.\n')
df = pd.DataFrame(rows, columns = columns)
df.to_excel('Eydens.xlsx')
print("File saved")
    
```
